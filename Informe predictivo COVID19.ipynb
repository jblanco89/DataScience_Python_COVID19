```python
inicio = pd.to_datetime("2020-02-01",format="%Y-%m-%d %H:%M:%S")
df['days'] = (today-inicio).dt.days
df
```

## Modelo Logísitico

$$ f(x,a,b,c) = \frac{c}{1+e^{-(x-b)/a}}$$

a = velocidad de infección
b = día de maxima infección
c = total de infectados registrados al final de la infección

En valores de tiempo altos, el número de personas infectadas se acerca cada vez más a c, y ese es el punto en el que podemos decir que la infección ha terminado. Esta función también tiene un punto de inflexión en b, que es el punto en el que la primera derivada comienza a disminuir (es decir, el pico después del cual la infección comienza a ser menos agresiva y disminuye).

### definiendo el modelo en python

```python
def logistic_model (x,a,b,c):
    return c/(1+np.exp(-(x-b)/a))
```

Para estimar los parametros a,b,c así como sus errores usaré el paquete `curve_fit` de `scipy`

```python
x = list(df.iloc[:,2])
y = list(df.iloc[:,1])

fit = curve_fit(logistic_model,x,y,p0=[10,50,53578])

fit
```

a =5.19;
b = 50.35;
c= 121717.16;

La función también devuelve la matriz de covarianza, cuyos valores diagonales son las varianzas de los parámetros. Tomando su raíz cuadrada podemos calcular los errores estándar.

```python
errors = [np.sqrt(fit[1][i][i]) for i in [0,1,2]]

errors
```

Podemos usar la función fsolve de scipy para encontrar numéricamente la raíz de la ecuación que define el día final de la infección.

```python
sol = int(fsolve(lambda x : logistic_model(x,5.19,50.35,121717.16) - int(121717.16),50.35))
sol
```

## Modelo Exponencial

$$f(x,a,b,c) = a*e^{b(x-c)}$$

```python
def exponential_model(x,a,b,c):
    return a*np.exp(b*(x-c))

exp_fit = curve_fit(exponential_model,x,y,p0 =[1,0,0])

exp_fit
```

```python
exp_errors = [np.sqrt(exp_fit[1][i][i]) for i in [0,1,2]]
exp_errors
```

```python
pred_x = list(range(max(x),sol))

plt.rcParams['figure.figsize'] = [14, 9]

plt.rc('font', size=14)
# Real data

plt.scatter(x,y,label="Real",color="red")
# Predicted logistic curve
plt.plot(x+pred_x, [logistic_model(i,fit[0][0],fit[0][1],fit[0][2]) for i in x+pred_x], label="Modelo Logístico" )
# Predicted exponential curve
plt.plot(x+pred_x, [exponential_model(i,exp_fit[0][0],exp_fit[0][1],exp_fit[0][2]) for i in x+pred_x], label="Modelo Exponencial" )
plt.legend()
plt.xlabel("Desde 24 de Febrero 2020")
plt.ylabel("Número total de infectados en el tiempo")
plt.ylim((min(y)*0.1,140000*1.1))

plt.show()
```

## Análisis de Residuales


Los residuos son las diferencias entre cada punto experimental y el punto teórico correspondiente. Podemos analizar los residuos de ambos modelos para verificar la mejor curva de ajuste. En una primera aproximación, cuanto menor sea el error cuadrático medio entre los datos teóricos y experimentales, mejor será el ajuste.

```python
y_pred_logistic = [logistic_model(i,fit[0][0],fit[0][1],fit[0][2]) 
                   for i in x]

y_pred_exp = [exponential_model(i,exp_fit[0][0],exp_fit[0][1],exp_fit[0][2]) 
              for i in x]


```

```python
mean_squared_error(y,y_pred_logistic)
```

```python
mean_squared_error(y,y_pred_exp)
```

# Conclusión

Aunque el modelo logístico parece ser el más razonable, la forma de la curva probablemente cambiará debido a efectos exógenos como nuevos puntos críticos de infección, acciones del gobierno para combatir la infección, etc.

Por eso creo que las predicciones de este modelo comenzarán a ser útiles solo en unas pocas semanas, razonablemente después del pico de infección.
