## Graficando datos obtenidos

Los ejes han sido ajustados con base en los datos más relevantes

```python
fig, axes = plt.subplots()
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Tahoma']
axes.bar(df['Fecha'],df['Confirmados'], label = "Confirmados", color = '#0B3948')
axes.bar(df['Fecha'],df['Recuperados'], label = "Recuperados",color = '#ACB0BD')
axes.bar(df['Fecha'],df['Fallecidos'], label = "Fallecidos", color = 'k')
plt.grid(False)
axes.set_facecolor('#E7ECEF')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim((min(df['Fallecidos'])*1.1,74000*1.1))
plt.xlim('2020-03-04', '2020-03-30')
axes.set_title ('COVID-2019 España', fontsize=16)
axes.set_xlabel('Fecha', fontsize = 14)
axes.set_ylabel('Total Casos' , fontsize = 14)
axes.legend(loc=2)
fig.set_size_inches(17,10)
```
